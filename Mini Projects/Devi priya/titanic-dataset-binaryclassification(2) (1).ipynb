```python
#importing modules
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
train_data=pd.read_csv("/kaggle/input/titanic/train.csv")
test_data=pd.read_csv("/kaggle/input/titanic/test.csv")
```

```python
train_data.head()
```

```python
train_data.tail()
```

```python
#statistics
train_data.describe()
```

```python
#PREPROCESSING
df=pd.concat([train_data,test_data])
df=df.reset_index()
```

```python
df.head()
```

```python
df.tail()
```

```python
#Finding null values
df.isnull().sum()
```

```python
#dropping unnecessay columns
df=df.drop("Cabin",axis=1)
```

```python
#filling the null values of age and Fare by their mean values
df["Age"]=df["Age"].fillna(df["Age"].mean())
df["Fare"]=df["Fare"].fillna(df["Fare"].mean())
```

```python
#filling the missing values in embarked by the mode as its is categorical data
df["Embarked"].mode()
```

```python
df["Embarked"]=df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
#info
df.info()
```

```python
#dropping unnecessary columns
df=df.drop(["Name","Ticket"],axis=1)
```

```python
df.info()
```

```python
df.head()
```

```python
#Description of Numerical Data
df.describe()
```

```python
#Description of Categorical Data
df.describe(include="object")
```

```python
#Handling Categorical Data
#Transforming categorical data into numerical data by ORDINAL ENCODING
from sklearn.preprocessing import OrdinalEncoder
categorical_columns=["Sex","Embarked"]
encoder=OrdinalEncoder()
encoder.fit(df[categorical_columns])
df[categorical_columns]=encoder.transform(df[categorical_columns])
```

```python
df.head()
```

```python
#splitting the data set for training and testing
trainlen=len(train_data)
testlen=len(test_data)
train=df.iloc[:trainlen,:]
test=df.iloc[trainlen:,:]
```

```python
train.head()
```

```python
test.head()
```

```python
#splitting
x=train.drop(["PassengerId","Survived"],axis=1)
```

```python
y=train["Survived"]
```

```python
x.head()
```

```python
y
```

```python
#Train,test,split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
print(x.shape)
print(y.shape)
```

```python
#TRAINING THE MODEL
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)

```

```python
#slope and intercept
print(lr.intercept_)
print(lr.coef_)
```

```python
test.head()
```

```python
x_test=test.drop(["PassengerId","Survived"],axis=1)
```

```python
x_test.head()
```

```python
#PREDICTION
predict=lr.predict(x_test)
predict
```
