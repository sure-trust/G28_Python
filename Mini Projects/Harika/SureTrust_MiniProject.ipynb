```python
class User:
    def __init__(self, username, password, name, email):
        self.username = username
        self.password = password
        self.name = name
        self.email = email
users = []
user_limit = 6
def user_login(username, password):
    for user in users:
        if user.username == username and user.password == password:
            return user
    return None
while True:
    print("Welcome to the User Login System")
    print(f"User Limit: {len(users)}/{user_limit}")
    if len(users) >= user_limit:
        print("User limit exceeded. Cannot add more users.")
        break
    print("Enter '1' to create a new user","\n"," '2' to login","\n"," '3' to display user details ","\n","'q' to quit: ")
    choice = input()
    if choice == '1':
        if len(users) < user_limit:
            username = input("Enter a username: ")
            password = input("Enter a password: ")
            name = input("Enter your name: ")
            email = input("Enter your email: ")
            users.append(User(username, password, name, email))
            print("User created successfully.")
        else:
            print("User limit exceeded. Cannot add more users.")
    elif choice == '2':
        username = input("Enter your username: ")
        password = input("Enter your password: ")
        user = user_login(username, password)
        if user:
            print(f"Login successful! Welcome, {user.name}")
        else:
            print("Login failed. Invalid username or password.")
    elif choice == '3':
        for user in users:
            print(f"Username: {user.username}","\n",f" Name: {user.name}","\n",f" Email: {user.email}")
    elif choice == 'q':
        break
    else:
        print("Invalid choice. Please try again.")
```

```python
print("Welcome",username,"!")
import csv
import pandas as pd
class Product:
    def __init__(self, product_id, price, product_name, base_color, gender, quantities):
        self.product_id = product_id
        self.price = price
        self.product_name = product_name
        self.base_color = base_color
        self.gender = gender
        self.quantities = quantities

class Cart:
    def __init__(self):
        self.items = []

    def add_item(self, product, quantity):
        if int(product.quantities) < quantity:
            raise Exception("Product out of stock")
        self.items.append((product, quantity))

    def get_cost(self):
        total_cost = 0
        for product, quantity in self.items:
            total_cost += float(product.price) * quantity
        return total_cost

def load_products_from_csv(csv_file_path):
    products = []
    with open(csv_file_path, "r") as f:
        reader = csv.reader(f)
        for row in reader:
            product_id, price, product_name, base_color, gender, quantities = row
            product = Product(product_id, price, product_name, base_color, gender, quantities)
            products.append(product)
    return products

def display_product_catalog(csv_file):
    df = pd.read_csv(csv_file)
    print(df.to_string())
display_product_catalog("product_samples.csv")

def add_product_to_cart(cart, product, quantity):
    cart.add_item(product, quantity)

def display_cart(cart):
    print("Cart:")
    for product, quantity in cart.items:
        print(f"{product.product_name} ({quantity}) - {product.price}")

def get_total_cost(cart):
    return cart.get_cost()

def place_order(cart):
    print("Order placed successfully!")

def main():
    products = load_products_from_csv("product_samples.csv")
    cart = Cart()
    display_product_catalog("product_samples.csv")
    product_id = input("Enter the product ID: ")
    quantity = int(input("Enter the quantity: "))

    try:
        add_product_to_cart(cart, products[int(product_id)], quantity)
    except IndexError:
        print("Invalid product ID")
        return
    except Exception as e:
        print(e)
        return

    display_cart(cart)

    total_cost = get_total_cost(cart)

    print(f"Total cost: {round(total_cost, 2)}")

    place_order_confirmation = input("Do you want to place an order? (y/n): ")

    while place_order_confirmation == "n":

      display_product_catalog("products_samples.csv")

      product_id = input("Enter the product ID: ")
      quantity = int(input("Enter the quantity: "))

      try:
          add_product_to_cart(cart, products[int(product_id)], quantity)
      except IndexError:
          print("Invalid product ID")
          continue
      except Exception as e:
          print(e)
          continue

      display_cart(cart)
      total_cost = get_total_cost(cart)

      place_order_confirmation = input("Do you want to place an order? (y/n): ")

    if place_order_confirmation == "y":
        place_order(cart)
        print(f"Total cost of {username}'s cart: {round(total_cost, 2)}")
    else:
        print("Thank you! Keep on shopping!")
        main()


if __name__ == "__main__":
    main()
```
