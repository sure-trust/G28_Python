```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/kaggle/input/default-of-credit-card-clients-dataset/UCI_Credit_Card.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
from sklearn.cluster import KMeans
```

```python
import warnings
warnings.simplefilter("ignore")
```

```python
interia=[]
for i in range(1,10):
    kmeans=KMeans(n_clusters=i).fit(data)
    interia.append(kmeans.inertia_)
```

min(interia),max(interia)

```python
plt.plot(range(1,10),interia,"-*")
plt.title("Elbow graph")
```

```python
kmeans=KMeans(n_clusters=4).fit(data)
```

```python
ypred=kmeans.predict(data)
```

```python
data.iloc[ypred==0,0].shape==data.iloc[ypred==0,1].shape
```

```python
centroids=kmeans.cluster_centers_
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(2)
```

```python
reduced_data=pd.DataFrame(pca.fit_transform(data),columns=["PCA1","PCA2"])
```

```python
reduced_data.head()
```

```python
kmeans=KMeans(4).fit(reduced_data)
ypred=kmeans.predict(reduced_data)
```

```python
plt.scatter(reduced_data.loc[ypred==0,"PCA1"],reduced_data.loc[ypred==0,"PCA2"],c="r",label="c1",s=50)
plt.scatter(reduced_data.loc[ypred==1,"PCA1"],reduced_data.loc[ypred==1,"PCA2"],c="g",label="c2",s=50)
plt.scatter(reduced_data.loc[ypred==2,"PCA1"],reduced_data.loc[ypred==2,"PCA2"],c="b",label="c3",s=50)
plt.scatter(reduced_data.loc[ypred==3,"PCA1"],reduced_data.loc[ypred==3,"PCA2"],c="y",label="c4",s=50)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c="black",label="C5",s=150)
```
